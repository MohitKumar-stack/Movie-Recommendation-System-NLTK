```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import json
```

```python
movies_data =pd.read_csv("tmdb_5000_credits.csv")
credits_data=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits_data.head(2)
```

```python
credits_data=credits_data.merge(movies_data,on="title")
```

```python
credits_data.head()
```

```python
credits_data.dropna(inplace=True)
```

```python
credits_data.head(1)
```

```python
credits_data=credits_data[["movie_id","title","cast","crew","genres","overview","keywords"]]
```

```python
credits_data.head()
```

```python
import ast
#  convert string of list in list
```

```python
def convert(str):
    l=[]
    for i in ast.literal_eval(str):
        l.append(i["name"])
    return l    
    
```

```python
credits_data["genres"]=credits_data["genres"].apply(convert)
credits_data["keywords"]=credits_data["keywords"].apply(convert)
```

```python
credits_data.head()
```

```python
credits_data["cast"][0]
```

```python
# we are only fetching top 3 cast of any mvie
def fetch_top3_crew(str):
    l=[]
    temp=0
    for i in ast.literal_eval(str):
        if temp<=3:
            l.append(i["character"])
            temp+=1
    return l    
```

```python
credits_data["cast"]=credits_data["cast"].apply(fetch_top3_crew)
```

```python
credits_data.head()
```

```python
credits_data["crew"][0]
```

```python
def cobvert(str):
    l=[]
    temp=0
    for i in ast.literal_eval(str):
        if temp<=1 and i["job"]=="Director":
            l.append(i["name"])
            temp+=1
    return l  
```

```python
credits_data["crew"]=credits_data["crew"].apply(cobvert)
```

```python
credits_data["crew"]
```

```python
credits_data.head()
```

```python
credits_data["overview"]=credits_data["overview"].apply(lambda x:x.split())
```

```python
credits_data["overview"]=credits_data["overview"].apply(lambda x:[i.replace(" ","") for i in x])
credits_data["cast"]=credits_data["cast"].apply(lambda x:[i.replace(" ","") for i in x])
credits_data["crew"]=credits_data["crew"].apply(lambda x:[i.replace(" ","") for i in x])
credits_data["keywords"]=credits_data["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
credits_data["genres"]=credits_data["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
credits_data.head(2)
```

```python
credits_data["Tags"]=credits_data["cast"]+credits_data["crew"]+credits_data["genres"]+credits_data["overview"]+credits_data["keywords"]
```

```python
new_credits_data =credits_data[["movie_id","title","Tags"]]
```

```python
new_credits_data.head()
```

```python
new_credits_data["Tags"]=new_credits_data["Tags"].apply(lambda x:" ".join(x))
```

```python
new_credits_data["Tags"]=new_credits_data["Tags"].apply(lambda x:x.lower())
```

```python
new_credits_data
```

```python
movies_list=list(new_credits_data["title"])
```

```python
movies_list
```

```python
with open('movies_list.json', 'w') as json_file:
    json.dump(movies_list, json_file)
```

```python
# import nltk to stem the data remove the same kind of word
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
    
```

```python
new_credits_data["Tags"]=new_credits_data["Tags"].apply(stem)
```

```python
CV=CountVectorizer(max_features=5000,stop_words="english")
```

```python
Vectors=CV.fit_transform(new_credits_data["Tags"]).toarray()
```

```python
CV.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(Vectors)
```

```python
# convert into json so that we can import this imto flask app file
with open('similarity.json', 'w') as json_file:
    json.dump(similarity.tolist(), json_file)
```

```python
def recommend(movie):
    movie_index=new_credits_data[new_credits_data["title"]==movie].index[0]
    distance =similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_credits_data.iloc[i[0]].title)
        
    
```

```python
recommend("Iron Man")
```
